```python
import pandas as pd
import numpy as np
```

# Public Hospital data

```python
data=pd.read_excel(r"E:\career fair\ghg data\normalized_bps_data_2011-2020_public_hospital.xlsx",sheet_name=['2011','2012','2013','2014','2015','2016','2017','2018','2019','2020'])
data_2011=data['2011']
data_2012=data['2012']
data_2013=data['2013']
data_2014=data['2014']
data_2015=data['2015']
data_2016=data['2016']
data_2017=data['2017']
data_2018=data['2018']
data_2019=data['2019']
data_2020=data['2020']
```

```python
data_2011['year']=2011
data_2012['year']=2012
data_2013['year']=2013
data_2014['year']=2014
data_2015['year']=2015
data_2016['year']=2016
data_2017['year']=2017
data_2018['year']=2018
data_2019['year']=2019
data_2020['year']=2020
```

```python
data_2011.rename(columns={'2011 GHG Emissions (kg)':'GHG Emissions','2011 Energy Intensity (eWh/HDD/sq. ft)':'Energy Intensity','2011 Energy Intensity (ekWh/ML)':'nothing'},inplace=True)
data_2012.rename(columns={'2012 GHG Emissions (kg)':'GHG Emissions','2012 Energy Intensity (eWh/HDD/sq. ft)':'Energy Intensity','2012 Energy Intensity (ekWh/ML)':'nothing'},inplace=True)
data_2013.rename(columns={'2013 GHG Emissions (kg)':'GHG Emissions','2013 Energy Intensity (eWh/HDD/sq. ft)':'Energy Intensity','2013 Energy Intensity (ekWh/ML)':'nothing'},inplace=True)
data_2014.rename(columns={'2014 GHG Emissions (kg)':'GHG Emissions','2014 Energy Intensity (eWh/HDD/sq. ft)':'Energy Intensity','2014 Energy Intensity (ekWh/ML)':'nothing'},inplace=True)
data_2015.rename(columns={'2015 GHG Emissions (kg)':'GHG Emissions','2015 Energy Intensity (eWh/HDD/sq. ft)':'Energy Intensity','2015 Energy Intensity (ekWh/ML)':'nothing'},inplace=True)
data_2016.rename(columns={'2016 GHG Emissions (kg)':'GHG Emissions','2016 Energy Intensity (eWh/HDD/sq. ft)':'Energy Intensity','2016 Energy Intensity (ekWh/ML)':'nothing'},inplace=True)
data_2017.rename(columns={'2017 GHG Emissions (kg)':'GHG Emissions','2017 Energy Intensity (eWh/HDD/sq. ft)':'Energy Intensity','2017 Energy Intensity (ekWh/ML)':'nothing'},inplace=True)
data_2018.rename(columns={'2018 GHG Emissions (kg)':'GHG Emissions','2018 Energy Intensity (eWh/HDD/sq. ft)':'Energy Intensity','2018 Energy Intensity (ekWh/ML)':'nothing'},inplace=True)
data_2019.rename(columns={'2019 GHG Emissions (kg)':'GHG Emissions','2019 Energy Intensity (eWh/HDD/sq. ft)':'Energy Intensity','2019 Energy Intensity (ekWh/ML)':'nothing'},inplace=True)
data_2020.rename(columns={'GHG Emissions (kg)':'GHG Emissions','Energy Intensity (eWh/HDD/sq. ft)':'Energy Intensity','Energy Intensity (ekWh/ML)':'nothing','SubSector':'Sub Sector'},inplace=True)
```

```python
data_2020.columns
```

# data concat

```python
hospital_data=pd.concat([data_2011,data_2012,data_2013,data_2014,data_2015,data_2016,data_2017,data_2018,data_2019,data_2020]).reset_index()
hospital_data.drop(columns=['nothing','index'],inplace=True)
```

```python
hospital_data
```

# null value

```python
hospital_data.drop(k.index,inplace=True,axis=0)
```
